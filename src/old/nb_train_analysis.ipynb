```python
import numpy as np
import matplotlib.pyplot as plt

drive = 'drive/MyDrive/data/audio2face/'
```

```python
target_file = drive + 'Emma_val_blendshape.txt'
# predicted_file = drive + 'Emma_LSTMNvidiaNet_hat.txt'
predicted_file = drive + 'Emma_NvidiaNet_hat.txt'

# best
# predicted_file = drive + 'best/Copy of Emma_LSTMNvidiaNet_hat.txt'
# predicted_file = drive + 'best/Copy of Emma_NvidiaNet_hat.txt'

y = np.loadtxt(target_file)
y_hat = np.loadtxt(predicted_file)
```

```python
y
```

# plot

```python
figure, axis = plt.subplots(nrows = 1, ncols = 2, figsize = (16, 6));

axis[0].imshow(y, aspect = 'auto', vmin = 0, vmax = 1, cmap = 'gray');
axis[0].set_title('y');
axis[1].imshow(y_hat, aspect = 'auto', vmin = 0, vmax = 1, cmap = 'gray');
axis[1].set_title('y_hat');
```

# plot vector

```python
figure, axis = plt.subplots(nrows = 2, ncols = 2, figsize = (32, 12));

axis[0, 0].plot(y[:, 2], label = 'y');
axis[0, 0].plot(y_hat[:, 2], label = 'y_hat');
axis[0, 0].set_title('[:, 2]');
axis[0, 0].legend();
axis[0, 1].plot(y[:, 3], label = 'y');
axis[0, 1].plot(y_hat[:, 3], label = 'y_hat');
axis[0, 1].set_title('[:, 3]');
axis[0, 1].legend();

axis[1, 0].plot(y[:, 30], label = 'y');
axis[1, 0].plot(y_hat[:, 30], label = 'y_hat');
axis[1, 0].set_title('[:, 30]');
axis[1, 0].legend();
axis[1, 1].plot(y[:, 32], label = 'y');
axis[1, 1].plot(y_hat[:, 32], label = 'y_hat');
axis[1, 1].set_title('[:, 32]');
axis[1, 1].legend();
```

```python
figure, axis = plt.subplots(nrows = 2, ncols = 2, figsize = (16, 12));

idx = np.argsort(y[:, 2])
axis[0, 0].plot(y[idx, 2], label = 'y');
axis[0, 0].plot(y_hat[idx, 2], label = 'y_hat');
axis[0, 0].set_title('[argsort(y), 2]');
axis[0, 0].legend();

idx = np.argsort(y[:, 3])
axis[0, 1].plot(y[idx, 3], label = 'y');
axis[0, 1].plot(y_hat[idx, 3], label = 'y_hat');
axis[0, 1].set_title('[argsort(y), 3]');
axis[0, 1].legend();

idx = np.argsort(y[:, 30])
axis[1, 0].plot(y[idx, 30], label = 'y');
axis[1, 0].plot(y_hat[idx, 30], label = 'y_hat');
axis[1, 0].set_title('[argsort(y), 30]');
axis[1, 0].legend();

idx = np.argsort(y[:, 32])
axis[1, 1].plot(y[idx, 32], label = 'y');
axis[1, 1].plot(y_hat[idx, 32], label = 'y_hat');
axis[1, 1].set_title('[argsort(y), 32]');
axis[1, 1].legend();
```

```python
figure, axis = plt.subplots(nrows = 2, ncols = 2, figsize = (16, 12));

# idx = np.argsort(y[:, 2])
axis[0, 0].plot(y[:, 2] - y_hat[:, 2], label = 'y - y_hat');
# axis[0, 0].plot(y_hat[idx, 2], label = 'y_hat');
axis[0, 0].set_title('[argsort(y), 2]');
axis[0, 0].legend();

idx = np.argsort(y[:, 3])
axis[0, 1].plot(y[idx, 3], label = 'y');
axis[0, 1].plot(y_hat[idx, 3], label = 'y_hat');
axis[0, 1].set_title('[argsort(y), 3]');
axis[0, 1].legend();

idx = np.argsort(y[:, 30])
axis[1, 0].plot(y[idx, 30], label = 'y');
axis[1, 0].plot(y_hat[idx, 30], label = 'y_hat');
axis[1, 0].set_title('[argsort(y), 30]');
axis[1, 0].legend();

idx = np.argsort(y[:, 32])
axis[1, 1].plot(y[idx, 32], label = 'y');
axis[1, 1].plot(y_hat[idx, 32], label = 'y_hat');
axis[1, 1].set_title('[argsort(y), 32]');
axis[1, 1].legend();
```
